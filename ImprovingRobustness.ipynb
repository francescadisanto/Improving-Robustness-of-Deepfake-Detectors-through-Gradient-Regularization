```python
#Extracts each zip file in the provided list into a folder inside extract_to.
def extract_all_zips(zip_files, extract_to="DFFD/raw"):
    """
    Args:
        zip_files (list): List of .zip file names in the working directory.
        extract_to (str): Destination folder where zip contents will be extracted.
    """
    os.makedirs(extract_to, exist_ok=True)

    for zip_name in zip_files:
        folder_name = zip_name.replace(".zip", "")
        full_path = os.path.join(extract_to, folder_name)

        print(f"Extracting {zip_name} into {extract_to}...")
        with zipfile.ZipFile(zip_name, 'r') as zip_ref:
            zip_ref.extractall(extract_to)

    print("All folders extracted correctly.")
```

```python
class DeepfakeBinaryDataset(Dataset):
    """
    Custom Dataset for binary classification of real vs fake faces.
    Supports multiple input directories, extension filtering,
    automatic corruption handling, and optional transform.
    """

    def __init__(self, real_dirs, fake_dirs, transform=None, extensions=(".jpg", ".jpeg", ".png")):
        self.samples = []
        self.transform = transform

        def valid_image(path):
            return path.lower().endswith(extensions)

        #load real samples (label = 0)
        for real_dir in real_dirs:
            real_images = glob(os.path.join(real_dir, "*"))
            self.samples.extend([(img, 0) for img in real_images if valid_image(img)])

        #load fake samples (label = 1)
        for fake_dir in fake_dirs:
            fake_images = glob(os.path.join(fake_dir, "*"))
            self.samples.extend([(img, 1) for img in fake_images if valid_image(img)])

        #shuffle samples
        random.shuffle(self.samples)

        #check if dataset is empty
        if len(self.samples) == 0:
            print("Warning: no images found in provided directories. Check your paths and extensions.")

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        img_path, label = self.samples[idx]
        try:
            image = Image.open(img_path).convert("RGB")
        except Exception as e:
            print(f"Corrupted image: {img_path} (label={label}), error: {e}")
            image = Image.new("RGB", (224, 224))  # fallback

        if self.transform:
            image = self.transform(image)

        return image, label
```

```python
#Model Baseline definition
LEARNING_RATE = 3e-5
NUM_EPOCHS = 10

model_baseline = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.IMAGENET1K_V1)
model_baseline.classifier[1] = nn.Linear(model_baseline.classifier[1].in_features, 2)
model_baseline = model_baseline.to(device)


#loss and optimizer
criterion = nn.CrossEntropyLoss(weight=class_weights)
optimizer = optim.Adam(model_baseline.parameters(), lr=LEARNING_RATE)

#training and validation
def train_one_epoch(model, dataloader):
    model.train()
    running_loss, correct, total = 0.0, 0, 0
    for inputs, labels in tqdm(dataloader, desc="Training"):
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * inputs.size(0)
        _, preds = torch.max(outputs, 1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)
    return running_loss / total, correct / total

def evaluate(model, dataloader):
    model.eval()
    running_loss, correct, total = 0.0, 0, 0
    all_preds, all_labels = [], []
    with torch.no_grad():
        for inputs, labels in tqdm(dataloader, desc="Validation"):
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            _, preds = torch.max(outputs, 1)
            running_loss += loss.item() * inputs.size(0)
            correct += (preds == labels).sum().item()
            total += labels.size(0)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    epoch_loss = running_loss / total
    epoch_acc = correct / total
    epoch_f1 = f1_score(all_labels, all_preds, average='macro')
    print(f" Val   Loss: {epoch_loss:.4f} | Acc: {epoch_acc:.4f} | F1: {epoch_f1:.4f}")
    return epoch_loss, epoch_acc, epoch_f1
```

Baseline with gradient regularization


```python
# Perturbation Injection Module
class PerturbationInjectionModule(nn.Module):
    def forward(self, x_norm, mu, sigma, delta_mu, delta_sigma):
        return x_norm * (sigma + delta_sigma) + (mu + delta_mu)

# Modified EfficientNet with explicit shallow/deep separation
class EfficientNetSplit(nn.Module):
    def __init__(self):
        super().__init__()
        base = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.IMAGENET1K_V1)
        self.shallow_features = base.features[:2]
        self.deep_features = base.features[2:]
        self.avgpool = base.avgpool
        in_features = base.classifier[1].in_features
        self.classifier = nn.Sequential(
            nn.Dropout(p=0.3),
            nn.Linear(in_features, 2)
        )
        self.injector = PerturbationInjectionModule()

        for param in self.shallow_features.parameters():
            param.requires_grad = False

        for m in self.shallow_features.modules():
            if isinstance(m, nn.BatchNorm2d):
                m.eval()

    def forward_features(self, x):
        x = normalize_batch(x)
        x = self.shallow_features(x)
        return x

    def forward_from_shallow(self, shallow):
        x = self.deep_features(shallow)
        x = self.avgpool(x)
        x = torch.flatten(x, 1)
        return self.classifier(x)

    def forward(self, x):
        shallow = self.forward_features(x)
        return self.forward_from_shallow(shallow)

# Compute statistics
def compute_statistics(x):
    mu = x.mean(dim=[2, 3], keepdim=True)
    sigma = torch.sqrt(x.var(dim=[2, 3], keepdim=True) + 1e-8)
    return mu, sigma

# Training with Gradient Regularization
def train_with_grad_reg(model, dataloader, optimizer, criterion, device, r_scalar, alpha_coeff, scaler=None):
    model.train()
    scaler = scaler or GradScaler()

    total_loss, correct, total = 0.0, 0, 0

    for inputs, labels in tqdm(dataloader, desc="Training (GR + PID)"):
        inputs, labels = inputs.to(device), labels.to(device)

        # Clean forward pass
        optimizer.zero_grad()
        with autocast():
            shallow = model.forward_features(inputs)
            mu, sigma = compute_statistics(shallow)
            mu = mu.detach().requires_grad_(True)
            sigma = sigma.detach().requires_grad_(True)
            x_norm = (shallow - mu) / (sigma + 1e-8)
            recon = model.injector(x_norm, mu, sigma, torch.zeros_like(mu), torch.zeros_like(sigma))
            out_clean = model.forward_from_shallow(recon)
            loss_clean = criterion(out_clean, labels)

        scaler.scale(loss_clean).backward(retain_graph=True)
        grad_mu, grad_sigma = torch.autograd.grad(loss_clean, [mu, sigma], retain_graph=True, create_graph=False)
        g1 = [p.grad.clone() if p.grad is not None else torch.zeros_like(p) for p in model.parameters()]

        # Perturbation step
        grad_vec = torch.cat([grad_mu.flatten(), grad_sigma.flatten()])
        norm = torch.norm(grad_vec)
        delta = r_scalar * grad_vec / norm if norm > 1e-8 else torch.zeros_like(grad_vec)
        delta_mu = delta[:mu.numel()].view_as(mu)
        delta_sigma = delta[mu.numel():].view_as(sigma)

        # Perturbed forward pass
        optimizer.zero_grad()
        with autocast():
            x_norm_2 = (shallow.detach() - mu.detach()) / (sigma.detach() + 1e-8)
            perturbed = model.injector(x_norm_2, mu.detach(), sigma.detach(), delta_mu, delta_sigma)
            out_perturbed = model.forward_from_shallow(perturbed)
            loss_perturbed = criterion(out_perturbed, labels)

        scaler.scale(loss_perturbed).backward()
        g2 = [p.grad.clone() if p.grad is not None else torch.zeros_like(p) for p in model.parameters()]

        # Update parameters
        optimizer.zero_grad()
        with torch.no_grad():
            for i, p in enumerate(model.parameters()):
                p.grad = (1 - alpha_coeff) * g1[i] + alpha_coeff * g2[i]

        scaler.step(optimizer)
        scaler.update()

        loss_combined = (1 - alpha_coeff) * loss_clean.item() + alpha_coeff * loss_perturbed.item()
        total_loss += loss_combined
        _, preds = torch.max(out_clean, 1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

    return total_loss / total, correct / total


def test_clean(model, test_loader):
    model.eval()
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for x, y in tqdm(test_loader, desc="Clean Test"):
            x = x.to(device)
            y = y.to(device)
            logits = model(x)
            preds = logits.argmax(dim=1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(y.cpu().numpy())
    acc = accuracy_score(all_labels, all_preds)
    f1 = f1_score(all_labels, all_preds, average="macro")
    print(f"\nClean Test - Acc: {acc:.4f} | F1: {f1:.4f}")
    return acc, f1

def evaluate_clean(model, dataloader, criterion, device):
    model.eval()
    running_loss, correct, total = 0.0, 0, 0
    all_preds, all_labels = [], []
    with torch.no_grad():
        for inputs, labels in tqdm(dataloader, desc="Validation"):
            inputs, labels = inputs.to(device), labels.to(device)
            out_clean = model(inputs)
            loss = criterion(out_clean, labels)
            running_loss += loss.item() * inputs.size(0)
            _, preds = torch.max(out_clean, 1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    acc = correct / total
    f1 = f1_score(all_labels, all_preds, average='macro')
    return running_loss / total, acc, f1
```

```python
#final plot
def plot_attack_comparison(epsilons, baseline_accs, reg_accs, baseline_f1s, reg_f1s, title):
    plt.figure(figsize=(8,5))
    plt.plot(epsilons, baseline_accs, marker='o', label='Baseline - Acc', color='blue')
    plt.plot(epsilons, reg_accs, marker='o', label='GR+PGD - Acc', linestyle='--', color='blue')
    plt.plot(epsilons, baseline_f1s, marker='s', label='Baseline - F1', color='green')
    plt.plot(epsilons, reg_f1s, marker='s', label='GR+PGD - F1', linestyle='--', color='green')
    plt.title(title)
    plt.xlabel("Epsilon")
    plt.ylabel("Score")
    plt.ylim(0, 1.05)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()


# table comparison
def print_comparison_table(epsilons, acc_base, acc_reg, f1_base, f1_reg):
    headers = ["Epsilon", "Acc Baseline", "Acc GR+PGD", "F1 Baseline", "F1 GR+PGD"]
    rows = [
        [f"{eps:.2f}", f"{a1:.4f}", f"{a2:.4f}", f"{f1:.4f}", f"{f2:.4f}"]
        for eps, a1, a2, f1, f2 in zip(epsilons, acc_base, acc_reg, f1_base, f1_reg)
    ]
    print("\n Comparison Table:")
    print(tabulate(rows, headers=headers, tablefmt="grid"))
```

```python
ablation_configs = {
    "config_4": {"alpha_coeff": 1.0, "r_scalar": 0.05},
    "config_5": {"alpha_coeff": 1.0, "r_scalar": 0.1},
    "config_6": {"alpha_coeff": 0.75, "r_scalar": 0.05},
    "config_7": {"alpha_coeff": 0.75, "r_scalar": 0.1}
}

# parameters
EPOCHS = 1
LEARNING_RATE = 3e-5

ablation_results = {}

#ablation loop
for name, cfg in ablation_configs.items():
    print(f"\n=== Start ablation: {name} (r = {cfg['r_scalar']}, α = {cfg['alpha_coeff']}) ===")


    torch.cuda.empty_cache()
    gc.collect()

    # Model
    model = EfficientNetSplit().to(device)
    optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)
    criterion = nn.CrossEntropyLoss()
    scaler = GradScaler()

    # Training
    for epoch in range(EPOCHS):
        print(f"\nEpoch {epoch+1}/{EPOCHS}")
        train_loss, train_acc = train_with_grad_reg(
            model=model,
            dataloader=train_loader,
            optimizer=optimizer,
            criterion=criterion,
            device=device,
            r_scalar=cfg["r_scalar"],
            alpha_coeff=cfg["alpha_coeff"],
            scaler=scaler
        )
        print(f"Train loss: {train_loss:.4f} | Train acc: {train_acc:.4f}")

    # Validation
    val_loss, val_acc, val_f1 = evaluate_clean(model, val_loader, criterion, device)
    print(f"Val acc: {val_acc:.4f} | Val f1: {val_f1:.4f}")

    # Save results
    ablation_results[name] = {
        "alpha": cfg["alpha_coeff"],
        "r_scalar": cfg["r_scalar"],
        "train_loss": train_loss,
        "train_acc": train_acc,
        "val_loss": val_loss,
        "val_acc": val_acc,
        "val_f1": val_f1
    }

    # Save model
    torch.save(model.state_dict(), f"models/reg_model_{name}.pth")

    # clean all
    del model, optimizer, criterion, scaler
    torch.cuda.empty_cache()
    gc.collect()

# final results
print("\n=== Results ABLATION STUDY ===")
for name, res in ablation_results.items():
    print(f"{name}: α={res['alpha']} | r={res['r_scalar']} | F1={res['val_f1']:.4f}")

# best config
best_config = max(ablation_results.items(), key=lambda x: x[1]['val_f1'])
best_name = best_config[0]
best_alpha = best_config[1]["alpha"]
best_r = best_config[1]["r_scalar"]
best_f1 = best_config[1]["val_f1"]

print(f"\nBest configuration: {best_name} → α = {best_alpha}, r = {best_r} | F1 = {best_f1:.4f}")
```

FGSM

```python
def fgsm_attack(image, label, model, epsilon):
    image = image.clone().detach().to(device).requires_grad_(True)
    label = label.to(device)

    output = model(normalize_batch(image))
    loss = F.cross_entropy(output, label)

    model.zero_grad()
    loss.backward()

    perturbed_image = image + epsilon * image.grad.sign()
    perturbed_image = torch.clamp(perturbed_image, 0, 1)

    return perturbed_image.detach()
```

```python
def test_fgsm(model, loader, epsilon, save_path=None):
    model.eval()
    correct = 0
    total = 0
    all_labels, all_preds, all_probs = [], [], []

    loop = tqdm(loader, desc=f"FGSM ε={epsilon:.4f}", leave=False)

    for images, labels in loop:
        images, labels = images.to(device), labels.to(device)

        adv_images = fgsm_attack(images, labels, model, epsilon)

        with torch.no_grad():
            outputs = model(normalize_batch(adv_images))
            probs = F.softmax(outputs, dim=1)[:, 1]
            preds = outputs.argmax(dim=1)

        correct += (preds == labels).sum().item()
        total += labels.size(0)

        all_labels.append(labels.cpu().numpy())
        all_preds.append(preds.cpu().numpy())
        all_probs.append(probs.cpu().numpy())

        loop.set_postfix(acc=correct / total)

    # Metrics
    all_labels = np.concatenate(all_labels)
    all_preds = np.concatenate(all_preds)
    all_probs = np.concatenate(all_probs)

    acc = correct / total
    f1 = f1_score(all_labels, all_preds, average='binary')

    print(f"FGSM ε={epsilon:.4f} | Accuracy: {acc:.4f} | F1: {f1:.4f}")

    if save_path:
        torch.save({
            "epsilon": epsilon,
            "labels": torch.tensor(all_labels),
            "preds": torch.tensor(all_preds),
            "probs": torch.tensor(all_probs),
        }, save_path)
        print(f"Saved attack output to {save_path}")

    return acc, f1, all_labels, all_preds, all_probs
```

```python
def visualize_attack_fgsm(model, dataloader, epsilon=0.05, n=4):
    model.eval()
    data_iter = iter(dataloader)
    orig_images, labels = next(data_iter)


    orig_images = orig_images[:n].to(device)
    labels = labels[:n].to(device)


    orig_images.requires_grad = True
    outputs = model(normalize_batch(orig_images))
    loss = criterion(outputs, labels)

    model.zero_grad()
    loss.backward()
    data_grad = orig_images.grad.data

    adv_images = orig_images + epsilon * data_grad.sign()
    adv_images = torch.clamp(adv_images, 0, 1)


    orig_images = orig_images.cpu().detach()
    adv_images = adv_images.cpu().detach()


    all_images = torch.cat([orig_images, adv_images], dim=0)
    grid = torchvision.utils.make_grid(all_images, nrow=n)
    npimg = grid.numpy().transpose((1, 2, 0))

    # Plot
    plt.figure(figsize=(16, 4))
    plt.imshow(npimg)
    plt.title(f"Original (top) vs FGSM ε={epsilon} (bottom)")
    plt.axis("off")
    plt.show()
```

PGD ATTACKS

```python
def pgd_attack(model, images, labels, epsilon, alpha, iters):
    images = images.clone().detach().to(device)
    labels = labels.to(device)
    ori_images = images.clone().detach()

    for i in range(iters):
        images.requires_grad_(True)
        outputs = model(images)
        loss = F.cross_entropy(outputs, labels)

        model.zero_grad()
        loss.backward()

        # Step in the direction of the gradient
        adv_images = images + alpha * images.grad.sign()

        # Project back to epsilon-ball and clip pixel values
        eta = torch.clamp(adv_images - ori_images, min=-epsilon, max=epsilon)
        images = torch.clamp(ori_images + eta, 0, 1).detach()

    return images
```

```python
def test_pgd(model, loader, epsilon, steps=5, alpha=None, save_path=None):
    model.eval()
    correct = 0
    total = 0
    all_labels, all_preds, all_probs = [], [], []

    alpha = alpha or (epsilon / steps)
    loop = tqdm(loader, desc=f"PGD ε={epsilon:.4f}", leave=False)

    for images, labels in loop:
        images, labels = images.to(device), labels.to(device)

        if epsilon == 0:
            adv_images = images.clone().detach()
        else:
            adv_images = pgd_attack(model, images, labels, epsilon, alpha, steps)

        with torch.no_grad():
            outputs = model(adv_images)
            probs = F.softmax(outputs, dim=1)[:, 1]
            preds = outputs.argmax(dim=1)

        correct += (preds == labels).sum().item()
        total += labels.size(0)

        all_labels.append(labels.cpu().numpy())
        all_preds.append(preds.cpu().numpy())
        all_probs.append(probs.cpu().numpy())

        loop.set_postfix(acc=correct / total)

    all_labels = np.concatenate(all_labels)
    all_preds = np.concatenate(all_preds)
    all_probs = np.concatenate(all_probs)

    acc = correct / total
    f1 = f1_score(all_labels, all_preds, average='binary')


    print(f"PGD ε={epsilon:.4f} | Accuracy: {acc:.4f} | F1: {f1:.4f}")

    if save_path:
        torch.save({
            "epsilon": epsilon,
            "labels": torch.tensor(all_labels),
            "preds": torch.tensor(all_preds),
            "probs": torch.tensor(all_probs),
        }, save_path)
        print(f"Saved attack output to {save_path}")

    return acc, f1,all_labels, all_preds, all_probs
```

MIFSGM ATTACK

```python
def mifgsm_attack(model, images, labels, epsilon, alpha, steps, decay=1.0):
    images = images.clone().detach().to(device)
    ori_images = images.clone().detach()
    labels = labels.to(device)

    momentum = torch.zeros_like(images).to(device)

    for i in range(steps):
        images.requires_grad_(True)
        outputs = model(images)
        loss = F.cross_entropy(outputs, labels)

        model.zero_grad()
        loss.backward()

        grad = images.grad.detach()
        grad_norm = grad / (torch.mean(torch.abs(grad), dim=(1,2,3), keepdim=True) + 1e-8)

        momentum = decay * momentum + grad_norm
        images = images + alpha * momentum.sign()
        images = torch.clamp(images, ori_images - epsilon, ori_images + epsilon)
        images = torch.clamp(images, 0, 1).detach()

    return images
```

```python
def test_mifgsm(model, loader, epsilon, steps=5, alpha=None, decay=1.0, save_path=None):
    model.eval()
    correct, total = 0, 0
    all_labels, all_preds, all_probs = [], [], []

    alpha = alpha or (epsilon / steps)
    loop = tqdm(loader, desc=f"MIFGSM ε={epsilon:.4f}", leave=False)

    for images, labels in loop:
        images, labels = images.to(device), labels.to(device)
        adv_images = mifgsm_attack(model, images, labels, epsilon, alpha, steps, decay)

        with torch.no_grad():
            outputs = model(adv_images)
            probs = F.softmax(outputs, dim=1)[:, 1]
            preds = outputs.argmax(dim=1)

        correct += (preds == labels).sum().item()
        total += labels.size(0)
        all_labels.append(labels.cpu().numpy())
        all_preds.append(preds.cpu().numpy())
        all_probs.append(probs.cpu().numpy())
        loop.set_postfix(acc=correct / total)

    all_labels = np.concatenate(all_labels)
    all_preds = np.concatenate(all_preds)
    all_probs = np.concatenate(all_probs)

    acc = correct / total
    f1 = f1_score(all_labels, all_preds, average='binary')

    print(f"MIFGSM ε={epsilon:.4f} | Accuracy: {acc:.4f} | F1: {f1:.4f}")

    if save_path:
        torch.save({
            "epsilon": epsilon,
            "labels": torch.tensor(all_labels),
            "preds": torch.tensor(all_preds),
            "probs": torch.tensor(all_probs),
        }, save_path)

    return acc, f1, all_labels, all_preds, all_probs
```

```python
def compute_gradcam(model, image, target_class, target_layer, device):
    model.eval()
    image = image.unsqueeze(0).to(device).requires_grad_()

    activations = []
    gradients = []

    def forward_hook(module, input, output):
        activations.append(output)

    def backward_hook(module, grad_input, grad_output):
        gradients.append(grad_output[0])

    handle_fwd = target_layer.register_forward_hook(forward_hook)
    handle_bwd = target_layer.register_backward_hook(backward_hook)

    output = model(normalize_batch(image))
    loss = F.cross_entropy(output, torch.tensor([target_class], device=device))
    model.zero_grad()
    loss.backward()

    grad = gradients[0][0]
    act = activations[0][0]
    weights = grad.mean(dim=(1, 2), keepdim=True)
    cam = (weights * act).sum(dim=0)
    cam = F.relu(cam)

    cam = cam - cam.min()
    cam = cam / (cam.max() + 1e-8)
    cam = cam.cpu().detach().numpy()
    cam = np.uint8(255 * cam)

    # Resize CAM and convert to 3-channel RGB
    cam_img = Image.fromarray(cam).resize((image.shape[3], image.shape[2]), resample=Image.BILINEAR)
    cam_img = np.array(cam_img)

    if cam_img.ndim == 2:
        cam_img = np.stack([cam_img]*3, axis=-1)  # (224,224,3)

    # Original image
    orig = image.squeeze().detach().cpu()
    if orig.shape[0] == 3:
        orig = orig.permute(1, 2, 0)
    orig = orig.numpy()
    orig = np.clip(orig, 0, 1)

    if cam_img.shape != orig.shape:
        print(f"Shape mismatch: cam_img {cam_img.shape}, orig {orig.shape}")
        cam_img = cv2.resize(cam_img, (orig.shape[1], orig.shape[0]))

    heatmap = plt.cm.jet(cam_img[:, :, 0] / 255.0)[:, :, :3]
    blended = 0.4 * heatmap + 0.6 * orig
    blended = np.uint8(blended * 255)

    handle_fwd.remove()
    handle_bwd.remove()

    return blended, cam_img
```

```python
# baseline: torchvision EfficientNet
model_baseline = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.IMAGENET1K_V1)
model_baseline.classifier[1] = nn.Linear(model_baseline.classifier[1].in_features, 2)
model_baseline = model_baseline.to(device)
model_baseline.load_state_dict(torch.load("models/baseline_efficientnet.pth"))
model_baseline.eval()

# regolarizzato: EfficientNetSplit
model_reg = EfficientNetSplit().to(device)
model_reg.load_state_dict(torch.load("models/reg_efficientnet.pth"))
model_reg.eval()

# Target layer for Grad-CAM
target_layer_baseline = model_baseline.features[-1]
target_layer_reg = model_reg.deep_features[-1]
```

```python
def save_attacked_gradcam(attack_name, attack_fn, epsilon, steps=3, alpha=None):
    os.makedirs(f"cam_inputs/{attack_name}", exist_ok=True)
    print(f"\nGenerating Grad-CAMs for {attack_name.upper()}...")

    real_count, fake_count = 0, 0
    max_per_class = 5
    img_id = 0

    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)

        for i in range(images.size(0)):
            x = images[i]
            y = labels[i]
            tag = "real" if y.item() == 0 else "fake"

            if (y.item() == 0 and real_count < max_per_class) or (y.item() == 1 and fake_count < max_per_class):
                x_in = x.unsqueeze(0)
                y_in = y.unsqueeze(0)

                if attack_name == "fgsm":
                    x_adv_base = attack_fn(x_in, y_in, model_baseline, epsilon).squeeze(0)
                    x_adv_reg = attack_fn(x_in, y_in, model_reg, epsilon).squeeze(0)
                elif attack_name == "pgd":
                    alpha = alpha if alpha is not None else epsilon / steps
                    x_adv_base = attack_fn(model_baseline, x_in, y_in, epsilon, alpha, steps).squeeze(0)
                    x_adv_reg = attack_fn(model_reg, x_in, y_in, epsilon, alpha, steps).squeeze(0)
                elif attack_name == "mifgsm":
                    alpha = alpha if alpha is not None else epsilon / steps
                    x_adv_base = attack_fn(model_baseline, x_in, y_in, epsilon, alpha, steps).squeeze(0)
                    x_adv_reg = attack_fn(model_reg, x_in, y_in, epsilon, alpha, steps).squeeze(0)
                else:
                    raise ValueError("Unsupported attack")

                cam_base, _ = compute_gradcam(model_baseline, x_adv_base, y.item(), target_layer_baseline, device)
                Image.fromarray(cam_base).save(f"cam_inputs/{attack_name}/baseline_img{img_id}_{tag}.png")

                cam_reg, _ = compute_gradcam(model_reg, x_adv_reg, y.item(), target_layer_reg, device)
                Image.fromarray(cam_reg).save(f"cam_inputs/{attack_name}/reg_img{img_id}_{tag}.png")

                if y.item() == 0:
                    real_count += 1
                else:
                    fake_count += 1
                img_id += 1

            if real_count >= max_per_class and fake_count >= max_per_class:
                break
        if real_count >= max_per_class and fake_count >= max_per_class:
            break

    print(f"Saved Grad-CAMs under cam_inputs/{attack_name}/")

# Esecuzione attacchi
save_attacked_gradcam("fgsm", fgsm_attack, epsilon=0.01)
save_attacked_gradcam("pgd", pgd_attack, epsilon=0.001, steps=5)
save_attacked_gradcam("mifgsm", mifgsm_attack, epsilon=0.001, steps=5)
```

```python
def compute_iinc(cam_path, image_tensor):
    cam = Image.open(cam_path).convert("RGB").resize((224, 224))
    cam_np = np.asarray(cam).astype(np.float32) / 255.0
    cam_gray = np.mean(cam_np, axis=2)

    image_np = image_tensor.permute(1, 2, 0).detach().cpu().numpy()
    image_np = np.clip(image_np, 0, 1)
    image_gray = np.mean(image_np, axis=2)

    return ssim(image_gray, cam_gray, data_range=1.0)

attack_list = ["fgsm", "pgd", "mifgsm"]
results = []

for attack in attack_list:
    path = f"cam_inputs/{attack}/"
    for fname in os.listdir(path):
        if fname.endswith(".png"):
            full_path = os.path.join(path, fname)
            model = "Baseline" if "baseline" in fname else "Regularized"
            tag = "real" if "real" in fname else "fake"
            img_id = int(fname.split("img")[1].split("_")[0])

            batch_idx = img_id // test_loader.batch_size
            sample_idx = img_id % test_loader.batch_size

            for b, (x_batch, _) in enumerate(test_loader):
                if b == batch_idx:
                    x = x_batch[sample_idx]
                    break

            iinc = compute_iinc(full_path, x)
            results.append({
                "Attack": attack.upper(),
                "Model": model,
                "Image_ID": img_id,
                "Class": tag,
                "IINC": iinc
            })

df_iinc = pd.DataFrame(results)
df_iinc.to_csv("results/iinc_scores.csv", index=False)

summary = df_iinc.groupby(["Attack", "Model"])["IINC"].mean().unstack()
summary_table = tabulate(summary.reset_index(), headers="keys", tablefmt="grid", showindex=False)
print(summary_table)


# Plot comparativo
attacks = summary.index.tolist()
baseline_vals = summary['Baseline'].tolist()
reg_vals = summary['Regularized'].tolist()

x = range(len(attacks))
width = 0.35

plt.figure(figsize=(8, 5))
plt.bar([p - width/2 for p in x], baseline_vals, width=width, label='Baseline')
plt.bar([p + width/2 for p in x], reg_vals, width=width, label='Regularized')
plt.xticks(x, attacks)
plt.ylabel("Mean IINC")
plt.title("Interpretability (IINC) Comparison per Attack")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
